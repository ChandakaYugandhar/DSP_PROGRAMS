# Pandas DataFrames

```python
import pandas as pd
data={
    "calories":[40,30,20],
    "energy":[55,33,55]
}
b=pd.DataFrame(data)
print(b)
```

```python
#loc is used to access the rows of the dataset
# [] is used to print in Dataframe
# without[] the result in series
print(b.loc[0])
print(b.loc[[0]])
```

# Named the indexes and Accesing


```python
import pandas as pd
data={
    "calories":[40,30,20],
    "energy":[55,33,55]
}
b=pd.DataFrame(data,index=["day 1","day 2","day 3"])
print(b)
```

```python
df=b.loc[["day 1"]]
print(df)
```
